#### <font color="cyan"><b>Operaci√≥n binaria</b></font>

<div class="alert alert-block alert-info">

<b>Definici√≥n (Operaci√≥n Binaria)</b>: <i>Una <b>operaci√≥n binaria</b> es aquella operaci√≥n matem√°tica, que necesita el operador y dos operandos (argumentos) para que se calcule un valor.</i>
    
<i>Dados tres conjuntos $A$, $B$ y $C$ una operaci√≥n binaria (<b>producto</b>), representando la operaci√≥n por el signo $\color{blue}{\circ}$, es una aplicaci√≥n que asigna a cada par de valores $a \in A$ y $b \in B$ un solo valor $c \in C$, que podemos representar:</i>


\begin{eqnarray*}
\circ :&A\times B&\longrightarrow &C\\
&(a,b)&\longmapsto &c
\end{eqnarray*}

<i>Podemos expresar la operaci√≥n:</i>

\begin{equation*}
    a\circ b = c, \quad\circ (a,b) = c, \quad(a,b)\xrightarrow \circ c
\end{equation*}

&(a,b)&\longmapsto &c = a + b

#### <font color="cyan"><b>Grupos</b></font>

<div class="alert alert-block alert-info">
<font color="blue" face="Ginebra">
    
<b>Definici√≥n (Grupo)</b>: <i>Un <b>grupo</b> $(G,\circ)$ es un conjunto $G$ en el que se ha definido una <b>operaci√≥n binaria interna</b> "$\circ$", que satisface los siguientes axiomas:</i>
    
<ul>
<li><i><b>Asociatividad:</b></i>
    
\begin{equation}\label{eq:G_Asoc}
    \forall\,\ a,b,c\in G:\,\ a\circ (b\circ c) = (a\circ b)\circ c
\end{equation}

<li><i><b>Elemento neutro:</b></i>
    
\begin{equation}\label{eq:G_Neut}
    \exists\,\ u\in G:\,\ u\circ a = a\circ u = a
\end{equation}

<li><i><b>Elemento inverso:</b></i>
    
\begin{equation}\label{eq:G_Inv}
    \forall\,\ a\in G \quad\exists\,\ a^{{-1}}\in G:\,\ a\circ a^{{-1}}=a^{{-1}}\circ a = u
\end{equation}

<li><i>Se dice que un grupo es <b>abeliano</b> o <b>conmutativo</b> cuando verifica adem√°s la <b>propiedad conmutativa:</b></i>
    
\begin{equation}\label{eq:G_Conm}
    a\circ b = b\circ a \quad\forall\,\ a,b\in G
\end{equation}
</ul>
    
</font>
</div>

#### <font color="cyan"><b>Campos</b></font>

<div class="alert alert-block alert-info">
<font color="blue" face="Ginebra">
    
<b>Definici√≥n (Campo)</b>: <i>Un <b>campo</b> $F$ es un conjunto en el cual se definen dos operaciones $+$ y $\cdot$ (llamadas, respectivamente, adici√≥n y multiplicaci√≥n) de modo que para cualquier par de elementos $a, b$ en $F$ existen elementos √∫nicos $a + b$ y $a \cdot b$ en $F$ tales que se cumplen las siguientes condiciones para todos los elementos $a, b, c$ en $F$:</i>

<ul>
<li><i><b>Conmutatividad de la adici√≥n y la multiplicaci√≥n.</b></i>
    
\begin{equation}\label{eq:F_Conm}
    a + b = b + a \quad\text{y} \quad\ a \cdot b = b \cdot a
\end{equation}

<li><i><b>Asociatividad de la adici√≥n y la multiplicaci√≥n.</b></i>
    
\begin{equation}\label{eq:F_Asoc}
    (a + b) + c = a + (b + c) \quad\text{y} \quad\ (a \cdot b) \cdot c = a \cdot (b \cdot c)
\end{equation}
    
<li><i><b>Existencia de elementos identidad para la adici√≥n y la multiplicaci√≥n.</b> Existen elementos distintos de $0$ y $1$ en $F$ tales que:</i></li>
    
\begin{equation}\label{eq:F_Iden}
    0 + a = a \quad\text{y} \quad\ 1 \cdot a = a
\end{equation}

<li><i><b>Existencia de inversos para la adici√≥n y la multiplicaci√≥n.</b> Para cada elemento $a$ en $F$ y cada elemento no nulo $b$ en $F$ existen elementos $c$ y $d$ en $F$ tales que:</i>
    
\begin{equation}\label{eq:F_Inv}
    a + c = 0 \quad\text{y} \quad\ b \cdot d = 1
\end{equation}
    
\begin{equation*}
    a + (-a) = 0 \quad\text{y} \quad\ b \cdot b^{-1} = b \cdot \frac{1}{b} = \frac{b}{b} = 1
\end{equation*}

<li><i><b>Distributividad de la multiplicaci√≥n sobre la adici√≥n:</b></i>
    
\begin{equation}\label{eq:F_Dist}
    a \cdot (b + c) = a \cdot b + a \cdot c
\end{equation}
</ul>
    
</font>
</div>

<div class="alert alert-block alert-danger">
<font color="purple" face="Ginebra">
    
<b>Ejemplos</b>:
    
<ul>
<li>$\mathbb{Z}$
    
<li>$\mathbb{R}$
    
<li>$\mathbb{C}$
</ul>
    
</font>
</div>

<div class="alert alert-block alert-success">
<font color="green" face="Ginebra">
    
<b>Teorema</b>: <i>Sean $a$ y $b$ elementos cualesquiera de un campo, entonces</i>
    
<ol>
<li>$a \cdot 0 = 0$

<li>$(-a)\cdot b = a \cdot (-b) = -(a \cdot b)$

<li>$(-a)\cdot (-b) = a \cdot b$
</ol>
    
</font>
</div>

<div class="alert alert-block alert-warning">
<font color="grey" face="Ginebra">
    
<b>Nota</b>: <i>Con los axiomas de campo y sus propiedades, podemos resolver diversos tipos de ecuaciones.</i>
    
</font>
</div>

<div class="alert alert-block alert-danger">
<font color="purple" face="Ginebra">

\begin{eqnarray*}
	ax + b &=& 0 \\
	(ax + b) \color{blue}{+ (-b)} &=& 0 \color{blue}{+ (-b)} &\quad\text{Sumamos el inverso aditivo de $b$} \\

</font>

<b>Nota</b>: <i>Trabajaremos en el campo de los n√∫meros reales, $\mathbb{R}$.</i>

<i>Se considera el problema de encontrar $n$ escalares (elementos de $\mathbb{R}$) $x_1, \dots, x_n$, que satisfagan las condiciones</i>

\begin{equation}\label{eq:sit_ecu_lin}
	\begin{matrix}
		a_{11}x_1 &+\,\cdots\,+& a_{1n}x_n & = & b_1 \\
		\vdots & & \vdots & = & \vdots \\
		a_{m1}x_1 &+\,\cdots\,+& a_{mn}x_n & = & b_m
	\end{matrix}
\end{equation}

<i>donde $b_1, \dots, b_n$ y $a_{ij}$ son elementos de $\mathbb{R}$, y se le llama <b>sistema de $m$ ecuaciones lineales con $n$ inc√≥gnitas</b>. El n√∫mero $n$ se conoce como el n√∫mero de <b>inc√≥gnitas</b> y $m$ es el n√∫mero de <b>ecuaciones</b>. Toda $n-$tupla $(x_1, \dots, x_n)$ de elementos de $\mathbb{R}$ que satisface cada una de las ecuaciones de \ref{eq:sit_ecu_lin}) se le llama <b>soluci√≥n</b> del sistema. Si $b_1 = b_2 = \cdots = b_m = 0$ se dice que el sistema es <b>homog√©neo</b>, o que cada una de las ecuaciones es homog√©nea.</i>
    
</font>
</div>

<div class="alert alert-block alert-danger">
<font color="purple" face="Ginebra">
    
<b>Ejemplo</b>: <i>Resolver el siguiente sistema de ecuaciones lineales</i>

\begin{eqnarray*}
	2x & + & 3y & = & 12 \\
	x & - & y & = & 1
\end{eqnarray*}

<b>Soluci√≥n</b>: <i>Tal vez, la t√©cnica fundamental para encontrar las soluciones de un sistema de ecuaciones lineales, es la t√©cnica de eliminaci√≥n.</i>

<i>Si sumamos ($-2$) veces la segunda ecuaci√≥n a la primera, se obtiene</i>

\begin{equation*}
	5y = 10
\end{equation*}

<i>De donde obtenemos que $y = 2$. Ahora, si sumamos $3$ veces la segunda ecuaci√≥n a la primera, se obtiene</i>

\begin{equation*}
	5x = 15
\end{equation*}

<i>De donde obtenemos que $x = 3$. Entonces la soluci√≥n de este sistema es la dupla $(3, 2)$.</i>

<b>Conclusi√≥n</b>: <i>Se ha encontrado la soluci√≥n del sistema ¬´eliminando inc√≥gnitas¬ª, esto es, multiplicando las ecuaciones por escalares y sum√°ndolas luego para producir ecuaciones en las cuales algunas de las variables no est√°n presentes.</i>
    
</font>
</div>

<div class="alert alert-block alert-warning">
<font color="grey" face="Ginebra">
    
<i>Desear√≠amos formalizar ligeramente este proceso de manera que se pueda entender por qu√© opera y para as√≠ poder llevar a cabo los c√°lculos necesarios para resolver un sistema de manera sistem√°tica.</i>
    
</font>
</div>

<i>Estaremos trabajando con <b>Wolfram Language</b>. En Wolfram el comando (o funci√≥n) que nos permite resolver una ecuaci√≥n o sitema de ecuaciones es: <b>Solve</b>, y su estructura es</i>

> $Solve[ecuaci√≥n, variable],\qquad\text{en el caso de una sola ecuaci√≥n}$

> $Solve[\{eq_1, \dots, eq_n\}, \{var_1, \dots, var_m\}],\qquad\text{en el caso de un sistema de } n \text{ ecuaciones con } m \text{ incognitas}$ 

```python
(*Escribimos nusteras ecuaciones*)
eq1 = 2*x + 3*y == 12;
eq2 = x - y == 1;
(*Resolvemos el sitema de ecuaciones con el comando Solve*)
Solve[{eq1, eq2}, {x, y}]
```

<div class="alert alert-block alert-info">
<font color="blue" face="Ginebra">
    
<i>Para el sistema general (\ref{eq:sit_ecu_lin}), suponga que seleccionamos $m$ escalares $c_1, \dots, c_m$, que se multiplica por la $i-$√©sima ecuaci√≥n por $c_i$ y luego se suma, es decir,</i>

\begin{equation*}
	\begin{matrix}
		c_1\big(a_{11}x_1 &+\,\cdots\,+& a_{1n}x_n\big) & = & c_1 b_1 \\
		\vdots & & \vdots & = & \vdots \\
		c_m\big(a_{m1}x_1 &+\,\cdots\,+& a_{mn}x_n\big) & = & c_m b_m
	\end{matrix}
\end{equation*}

<i>Sumando y agrupando tenemos</i>

\begin{equation*}
	\big(c_1 a_{11} + \cdots + c_m a_{m1}\big)x_1 + \cdots + \big(c_1 a_{1n} + \cdots + c_m a_{mn}\big)x_m
	= c_1 b_1 + \cdots + c_m b_m.
\end{equation*}

<i>Esta √∫ltima ecuaci√≥n se le llama <b>combinaci√≥n lineal</b> de las ecuaciones \ref{eq:sit_ecu_lin}). Observe que, cualquier soluci√≥n del sistema de ecuaciones \ref{eq:sit_ecu_lin}) ser√° tambi√©n soluci√≥n de esta nueva ecuaci√≥n. Esta es la idea fundamental del proceso de eliminaci√≥n.</i>
    
</font>
</div>

<div class="alert alert-block alert-info">
<font color="blue" face="Ginebra">
    
<i>Si se tiene otro sistema de ecuaciones lineales</i>

\begin{equation}\label{eq:sit_ecu_lin2}
	\begin{matrix}
		\alpha_{11}x_1 &+\,\cdots\,+& \alpha_{1n}x_n & = & \beta_1 \\
		\vdots & & \vdots & = & \vdots \\
		\alpha_{k1}x_1 &+\,\cdots\,+& \alpha_{kn}x_n & = & \beta_k
	\end{matrix}
\end{equation}

<i>en que cada una de las $K$ ecuaciones sea combinaci√≥n lineal de las ecuaciones de (\ref{eq:sit_ecu_lin}), entonces toda soluci√≥n de (\ref{eq:sit_ecu_lin}) es soluci√≥n de este nuevo sistema. Evidentemente puede suceder que algunas soluciones de (\ref{eq:sit_ecu_lin2}) no sean soluciones de (\ref{eq:sit_ecu_lin}). Esto, claro est√°, no sucede si cada ecuaci√≥n en el sistema original es combinaci√≥n lineal de las ecuaciones en el nuevo sistema. Se dir√° que dos sistemas de ecuaciones lineales son <b>equivalentes</b> si cada ecuaci√≥n de cada sistema es combinaci√≥n lineal de las ecuaciones del otro sistema.</i>
    
</font>
</div>

<div class="alert alert-block alert-success">
<font color="green" face="Ginebra">
    
<b>Teorema</b>: <i>Sistemas equivalentes de ecuaciones lineales tienen exactamente las mismas soluciones</i>
        
</font>
